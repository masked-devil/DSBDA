```python
import pandas as pd
```

```python
a=pd.read_csv(r"C:\Users\91935\Downloads\airquality.csv")
```

```python
#data cleaning
```

```python
a.isnull().sum()
```

```python
a=a.drop(columns=['Unnamed: 0'])
```

```python
a['Ozone']=a['Ozone'].fillna(a['Ozone'].mean())
a['Solar.R']=a['Solar.R'].fillna(a['Solar.R'].mean())
a['Humidity']=a['Humidity'].fillna(a['Humidity'].mode()[0])
```

```python
a
```

```python
a.isnull().sum()
```

```python
#label encoding
```

```python
from sklearn import preprocessing
a['Humidity']=preprocessing.LabelEncoder().fit_transform(a['Humidity'])
```

```python
a
```

```python
#data visualization
```

```python
import matplotlib.pyplot as plt
```

```python
#pie chart
```

```python
labels=['Ozone','Solar.R','Wind','Temperature']
sizes=[a['Ozone'].mean(),a['Solar.R'].mean(),a['Wind'].mean(),a['Temp'].mean()]
colors=['red','pink','yellow','green']
explode=(0.1,0,0,0)
plt.pie(sizes,explode,labels,colors,autopct='%1.1f%%',shadow=True)
plt.title('Average data')
plt.show()
```

```python
#bar chart
```

```python
import numpy as np
h=a.iloc[1:21,3]
y_pos=np.arange(len(h))
v=range(1,21)
plt.bar(y_pos,h)
plt.xticks(y_pos,v)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title('Temperatures for 1st to 20th May')
plt.show()
```

```python
#line graph
```

```python
h=a.iloc[1:16,0]
v=a.iloc[1:16,2]
plt.plot(h,label="Ozone",marker='*',linestyle='dotted')
plt.plot(v,label="Wind",marker='o',linestyle='dashed')
plt.legend()
plt.ylim(0,60)
plt.title("Ozone and Wind chart for 1st to 15th May")
plt.ylabel("Value Range")
plt.xlabel("Days")
plt.show()
```

```python
#histogram
```

```python
h=a.iloc[:,3]
plt.hist(h,bins='auto')
plt.title("Histogram for Temperatures")
plt.xlabel("No. of values")
plt.ylabel("Temperature")
plt.show()
```
